```
df = pd.read_csv("Sample_Data_Set.csv")
df
```

```
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn import preprocessing
from sklearn.svm import SVC
```

```
npdf = df.to_numpy()
npdf = np.delete(npdf, 16, 1)
npdf = np.delete(npdf, 15, 1)
npdf = np.delete(npdf, 13, 1)
npdf = np.delete(npdf, 0, 1)
npdf = np.delete(npdf, 1, 1)
npdf = np.delete(npdf, 1, 1)
npdf = np.delete(npdf, 1, 1)
npdf = np.delete(npdf, 1, 1)
npdf = np.delete(npdf, 2, 1)
npdf = np.delete(npdf, 2, 1)
npdf = np.delete(npdf, 2, 1)
npdf = npdf[:14649]
df2 = pd.DataFrame(npdf, columns=['0','1','2','3','4','5'])
df2
```

```
cv = CountVectorizer()

X = df2.loc[: ,"0"]
y = df2.loc[: ,"5"]
X_train, X_test, y_train, y_test = train_test_split(X.astype(str), y, test_size = 0.25, random_state = 123)
cv = CountVectorizer().fit(X_train)
df_train0 = pd.DataFrame(cv.fit_transform(X_train).todense(),columns = cv.get_feature_names())
df_test0 = pd.DataFrame(cv.transform(X_test).todense(), columns = cv.get_feature_names())

X = df2.loc[: ,"1"]
y = df2.loc[: ,"5"]
X_train, X_test, y_train, y_test = train_test_split(X.astype(str), y, test_size = 0.25, random_state = 123)
cv = CountVectorizer().fit(X_train)
df_train1 = pd.DataFrame(cv.transform(X_train).todense(),columns = cv.get_feature_names())
df_test1 = pd.DataFrame(cv.transform(X_test).todense(), columns = cv.get_feature_names())

X = df2.loc[: ,"2"]
y = df2.loc[: ,"5"]
X_train, X_test, y_train, y_test = train_test_split(X.astype(str), y, test_size = 0.25, random_state = 123)
cv = CountVectorizer().fit(X_train)
df_train2 = pd.DataFrame(cv.transform(X_train).todense(),columns = cv.get_feature_names())
df_test2 = pd.DataFrame(cv.transform(X_test).todense(), columns = cv.get_feature_names())

X = df2.loc[: ,"3"]
y = df2.loc[: ,"5"]
X_train, X_test, y_train, y_test = train_test_split(X.astype(str), y, test_size = 0.25, random_state = 123)
cv = CountVectorizer().fit(X_train)
df_train3 = pd.DataFrame(cv.transform(X_train).todense(),columns = cv.get_feature_names())
df_test3 = pd.DataFrame(cv.transform(X_test).todense(), columns = cv.get_feature_names())

X = df2.loc[: ,"4"]
y = df2.loc[: ,"5"]
X_train, X_test, y_train, y_test = train_test_split(X.astype(str), y, test_size = 0.25, random_state = 123)
cv = CountVectorizer().fit(X_train)
df_train4 = pd.DataFrame(cv.transform(X_train).todense(),columns = cv.get_feature_names())
df_test4 = pd.DataFrame(cv.transform(X_test).todense(), columns = cv.get_feature_names())
```

```
train = pd.concat([df_train0, df_train1, df_train2, df_train3, df_train4], axis = 1)
test = pd.concat([df_test0, df_test1, df_test2, df_test3, df_test4], axis = 1)
```

```
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, f1_score 

for i in [.01,.1,.5,1,10]:
  model = MultinomialNB(alpha = i)
  model.fit(train, y_train)
  y_predicted = model.predict(test)
  print("Alpha: ", i)
  print(accuracy_score(y_test, y_predicted))
```

```
from sklearn.ensemble import RandomForestClassifier
for i in [1,10,20,100,500]:
  model = RandomForestClassifier(random_state=0, n_estimators=i)
  model.fit(train, y_train)
  y_predicted = model.predict(test)
  print("N_estimator: ", i)
  print(accuracy_score(y_test, y_predicted))
```
